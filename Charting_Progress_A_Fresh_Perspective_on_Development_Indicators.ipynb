```python
#Cleaning the Donor data
# Renaming multiple columns
Donors.rename(columns={'T33': 'Region/Country/Area','Unnamed: 2': 'Year','Unnamed: 3': 'Series','Unnamed: 4':'Value (Million USD)','Unnamed: 5':'Footnotes','Unnamed: 6':'Source'}
          , inplace=True)
#Deleting the first Row
index_to_drop=0
Donors.drop(index_to_drop, inplace=True)
```

```python
Donors['Year'] = pd.to_numeric(Donors['Year'], errors='coerce')
Donors['Value (Million USD)'] = pd.to_numeric(Donors['Value (Million USD)'], errors='coerce')
```

```python
#Cleaning the Recipient data
# Renaming multiple columns
Recipient.rename(columns={'T32': 'Region/Country/Area','Unnamed: 2': 'Year','Unnamed: 3': 'Series','Unnamed: 4':'Value (Million USD)','Unnamed: 5':'Footnotes','Unnamed: 6':'Source'}
          , inplace=True)
#Deleting the first Row
index_to_drop=0
Recipient.drop(index_to_drop, inplace=True)
```

```python
Recipient['Year'] = pd.to_numeric(Recipient['Year'], errors='coerce')
Recipient['Value (Million USD)'] = pd.to_numeric(Recipient['Value (Million USD)'], errors='coerce')
```

Top Donors Over the Years.

```python
# Drop rows with NaN values in 'Value (Million USD)' column
Donors = Donors.dropna(subset=['Value (Million USD)'])
# Group the data by donor and year, and aggregate the net disbursements
grouped = Donors.groupby(['Net disbursements of official development assistance from donors', 'Year'])['Value (Million USD)'].sum().reset_index()

# Sum the net disbursements over the years for each donor
donor_totals = grouped.groupby('Net disbursements of official development assistance from donors')['Value (Million USD)'].sum()

# Sort the donors based on the total net disbursements
top_donors = donor_totals.sort_values(ascending=False).head(20)  # Selecting top 10 donors

# Define colors with varying shades of orange
colors = sns.color_palette("Oranges_r", len(top_donors))

# Plotting the bar graph with varying shades of orange
plt.figure(figsize=(10, 6))
sns.barplot(x=top_donors.index, y=top_donors.values, hue=top_donors.index, palette=colors, legend=False)
plt.title('Top Donors Over the Years')
plt.xlabel('Donor')
plt.ylabel('Total Net Disbursements (Million USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Drop rows with NaN values in 'Value (Million USD)' column
Recipient = Recipient.dropna(subset=['Value (Million USD)'])

# Group the data by recipient and year, and aggregate the net disbursements
grouped = Recipient.groupby(['Net disbursements of official development assistance to recipients', 'Year'])['Value (Million USD)'].sum().reset_index()

# Sum the net disbursements over the years for each recipient
recipient_totals = grouped.groupby('Net disbursements of official development assistance to recipients')['Value (Million USD)'].sum()

# Sort the recipients based on the total net disbursements
top_recipients = recipient_totals.sort_values(ascending=False).head(20)  # Selecting top 10 recipients

# Define colors with varying shades of orange
colors = sns.color_palette("Oranges_r", len(top_recipients))

# Plotting the bar graph with varying shades of orange
plt.figure(figsize=(10, 6))
sns.barplot(x=top_recipients.index, y=top_recipients.values, hue=top_recipients.index, palette=colors, legend=False)
plt.title('Top Recipients of Official Development Assistance Over the Years')
plt.xlabel('Recipient')
plt.ylabel('Total Net Disbursements (Million USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import plotly.graph_objs as go

# Group and aggregate the data by year for both donors and recipients
donors_grouped = Donors.groupby('Year')['Value (Million USD)'].sum().reset_index()
recipients_grouped = Recipient.groupby('Year')['Value (Million USD)'].sum().reset_index()

# Merge the aggregated data into a single DataFrame
merged_df = pd.merge(donors_grouped, recipients_grouped, on='Year', suffixes=('_donors', '_recipients'))

# Create traces for donors and recipients
donors_trace = go.Scatter(x=merged_df['Year'], y=merged_df['Value (Million USD)_donors'],
                          mode='lines', name='Donors', line=dict(color='blue'))
recipients_trace = go.Scatter(x=merged_df['Year'], y=merged_df['Value (Million USD)_recipients'],
                              mode='lines', name='Recipients', line=dict(color='red'))

# Create frames for animation
frames = [dict(data=[dict(type='scatter', x=[merged_df['Year'].iloc[i]], y=[merged_df['Value (Million USD)_donors'].iloc[i]]),
                      dict(type='scatter', x=[merged_df['Year'].iloc[i]], y=[merged_df['Value (Million USD)_recipients'].iloc[i]])],
               name=str(merged_df['Year'].iloc[i])) for i in range(len(merged_df))]

# Configure layout
layout = dict(title='Total Donors vs. Recipients Over the Years', xaxis=dict(title='Year'),
              yaxis=dict(title='Total Amount (Million USD)'), template='plotly_dark',
              updatemenus=[dict(type='buttons', showactive=False,
                                buttons=[dict(label='Play',
                                              method='animate',
                                              args=[None, dict(frame=dict(duration=500, redraw=False),
                                                               fromcurrent=True)])])])

# Create figure object
fig = go.Figure(data=[donors_trace, recipients_trace], layout=layout, frames=frames)

# Display animation
fig.show()
```
