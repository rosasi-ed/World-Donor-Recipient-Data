## <font face="courier" size="4"><font color ='salmon'> CLEANING HEADERS
<font face="courier" size="4">The first steps will ensure that the headings represents the data.

```python
column_mapping = {'T33': 'Region/Country/Area','Unnamed: 2': 'Year','Unnamed: 3': 'Series','Unnamed: 4': 'Value','Unnamed: 5': 'Footnotes','Unnamed: 6': 'Source'}
df.rename(columns=column_mapping, inplace=True)
column_mapping1 = {'T32': 'Region/Country/Area','Unnamed: 2': 'Year','Unnamed: 3': 'Series','Unnamed: 4': 'Value','Unnamed: 5': 'Footnotes','Unnamed: 6': 'Source'}
df1.rename(columns=column_mapping1, inplace=True)
```

```python
# Specify the index of the row you want to delete
index_to_delete = 0  # Change this to the index of the row you want to delete

# Delete the specified row
df = df.drop(index_to_delete)
df1 = df1.drop(index_to_delete)

# Resetting the index after deletion (optional)
df = df.reset_index(drop=True)
df1 = df1.reset_index(drop=True)

# Display the DataFrame after deleting a row
df.head(5)
```

```python
df1.head(5)
```

```python
column_mapping = {'Year': 'Year Donor','Series': 'Series Donor','Value': 'Value Donor(Million USD)','Footnotes': 'Footnotes Donor','Source': 'Source Donor'}
df.rename(columns=column_mapping, inplace=True)
column_mapping1 = {'Year': 'Year Recipient','Series': 'Series Recipient','Value': 'Value Recipient(Million USD)','Footnotes': 'Footnotes Recipient','Source': 'Source Recipient'}
df1.rename(columns=column_mapping1, inplace=True)
```

#<font face="courier" size="4"><font color='darksalmon'>**COMBINING CSVs**

```python
# Merge or concatenate the two DataFrames based on the common column
merged_df = pd.merge(df, df1, on='Region/Country/Area', how='outer')
# Change 'outer' to 'inner' if you want only common rows
# Display the merged DataFrame
# Save the merged DataFrame to a new CSV file
merged_df.to_csv('merged_file.csv', index=False)
```

```python
merged_df.head(5)
```

```python
merged_df.info()
```

```python
columns_to_convert = ['Value Donor(Million USD)', 'Value Recipient(Million USD)']
for col in columns_to_convert:
    merged_df[col] = pd.to_numeric(merged_df[col].str.replace(',', ''), errors='coerce', downcast='integer')
    merged_df[col] = merged_df[col].fillna(0).astype(int)
```

```python
merged_df.head(5)
```

```python
# List of columns to drop
columns_to_drop = ['Footnotes Donor', 'Footnotes Recipient']
# Drop the specified columns
merged_df = merged_df.drop(columns=columns_to_drop, errors='ignore')
```
